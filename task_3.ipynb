```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# read the data_clean.csv file
df=pd.read_csv('data_clean.csv')
df.head()
```

```python
df.info()
```

```python
df=df.drop(['Observer'], axis=1)
df.head()
```

```python
df=df.drop(['Year'], axis=1)
df.head()
```

Year and Observer not effect to "Yield" or "Crop Damage"

# Data Modelling

### (a) “Yield” is target y 

```python
x=df.drop('Yield',axis=1)
x.sample(5)
```

### Data PreProsessing

```python
# Replace 'Yes' with True and 'No' with False in columns
x['Heat Wave'] = x['Heat Wave'].map({'Y': True, 'N': False})
x['Dry Spell'] = x['Dry Spell'].map({'Y': True, 'N': False})
x['Cold Wave'] = x['Cold Wave'].map({'Y': True, 'N': False})
x['Wet Spell'] = x['Wet Spell'].map({'Y': True, 'N': False})
x['Irrigation'] = x['Irrigation'].map({'Y': True, 'N': False})
x['Crop Damage'] = x['Crop Damage'].map({'Y': True, 'N': False})
```

```python
x.head()
```

```python
x=pd.get_dummies(x)
x.head()
```

```python
y=df['Yield']
y.sample(5)
```

### Save As .csv

```python
x_train, x_test, y_train, y_test =train_test_split(x,y,test_size=0.2,random_state=5)
```

#### Training data (80%) 

```python
x_train.to_csv('a_X_train.csv',index=False)
x_train.head()
```

```python
y_train.to_csv('a_Y_train.csv',index=False)
y_train.head()
```

#### Testing data (20%)

```python
x_test.to_csv('a_X_test.csv',index=False)
x_test.head()
```

```python
y_test.to_csv('a_Y_test.csv',index=False)
y_test.head()
```
